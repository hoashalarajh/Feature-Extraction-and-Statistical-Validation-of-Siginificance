```python
# decision making function for t-test
def decision_making(result):
    p_val = result[-1]
    #making decison and interpretation
    if p_val < 0.05:
        print ("Null hypthesis of 'means of sample1 and sample2 are equal' is rejected")
        print ("There is enough evidence to show that there is a statistically significant difference in the means of sample1 and sample2\n")
    else:
        print ("Couldnt reject the Null hypthesis of 'means of sample1 and sample2 are equal'")
        print ("There is no statistically significant evidence to show 'means of sample1 and sample2 are not equal'\n")
```

### Conducting two-sample t-test

```python
# using scipy to conduct two-sampled t-test
import scipy.stats as stats
 
# two sampled t-test for two independent samples without equal variances
result1 = stats.ttest_ind(idle_face_dist_1, speaking_face_dist_1, equal_var=False)
print (result1)
decision_making(result1)

result2 = stats.ttest_ind(idle_face_dist_2, speaking_face_dist_2, equal_var=False)
print (result2)
decision_making(result2)

result3 = stats.ttest_ind(idle_face_dist_3, speaking_face_dist_3, equal_var=False)
print (result3)
decision_making(result3)

result4 = stats.ttest_ind(idle_face_dist_4, speaking_face_dist_4, equal_var=False)
print (result4)
decision_making(result4)

result5 = stats.ttest_ind(idle_face_dist_5, speaking_face_dist_5, equal_var=False)
print (result5)
decision_making(result5)

result6 = stats.ttest_ind(idle_face_dist_6, speaking_face_dist_6, equal_var=False)
print (result6)
decision_making(result6)

result7 = stats.ttest_ind(idle_face_dist_7, speaking_face_dist_7, equal_var=False)
print (result7)
decision_making(result7)

result8 = stats.ttest_ind(idle_face_dist_8, speaking_face_dist_8, equal_var=False)
print (result8)
decision_making(result8)

result9 = stats.ttest_ind(idle_face_dist_9, speaking_face_dist_9, equal_var=False)
print (result9)
decision_making(result9)

result10 = stats.ttest_ind(idle_face_dist_10, speaking_face_dist_10, equal_var=False)
print (result10)
decision_making(result10)

result11 = stats.ttest_ind(idle_face_dist_11, speaking_face_dist_11, equal_var=False)
print (result11)
decision_making(result11)

result12 = stats.ttest_ind(idle_face_dist_12, speaking_face_dist_12, equal_var=False)
print (result12)
decision_making(result12)

result13 = stats.ttest_ind(idle_face_dist_13, speaking_face_dist_13, equal_var=False)
print (result13)
decision_making(result13)

result14 = stats.ttest_ind(idle_face_dist_14, speaking_face_dist_14, equal_var=False)
print (result14)
decision_making(result14)

result15 = stats.ttest_ind(idle_face_dist_15, speaking_face_dist_15, equal_var=False)
print (result15)
decision_making(result15)

result16 = stats.ttest_ind(idle_face_dist_16, speaking_face_dist_16, equal_var=False)
print (result16)
decision_making(result16)
```
